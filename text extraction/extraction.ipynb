```python
from PyPDF2 import PdfReader

def extract_text_pypdf2(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Example usage
pdf_path1 = "C:/Users/moham/Downloads/DATA/cravath-bringing-blockchain-due-diligence-into-focus-102024_vb-002.pdf"
pdf_path2 = "C:/Users/moham/Downloads/DATA/DASCPWhitePaper.pdf"
pdf_path3 = "C:/Users/moham/Downloads/DATA/guide-to-the-vetting-of-digital-asset-and-digital-asset-exchanges.pdf"
textpdf1 = extract_text_pypdf2(pdf_path1)
textpdf2 = extract_text_pypdf2(pdf_path2)
textpdf3 = extract_text_pypdf2(pdf_path3)
print(textpdf1)
print(textpdf2)
print(textpdf3)
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

# Download NLTK resources (run once)
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove special characters, numbers, and extra whitespace
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Preprocess the extracted text
processed_text1 = preprocess_text(textpdf1)
processed_text2 = preprocess_text(textpdf2)
processed_text3 = preprocess_text(textpdf3)
```

```python
def save_text(text, output_path):
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(text)

# Save raw text
save_text(textpdf1, "cravath_raw.txt")
save_text(textpdf2, "dascp_raw.txt")
save_text(textpdf3, "vetting_guide_raw.txt")

# Save preprocessed tokens
save_text(' '.join(processed_text1), "cravath_processed.txt")
save_text(' '.join(processed_text2), "dascp_processed.txt")
save_text(' '.join(processed_text3), "vetting_guide_processed.txt")
```
