```python


# Configuration
MODEL_ID = "stabilityai/stable-diffusion-2-1"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
EMBEDDING_MODEL = "multi-qa-mpnet-base-dot-v1"
MISTRAL_MODEL = "meta-llama/llama-4-maverick:free"
FAISS_INDEX_PATH = r"G:\Ela\study\4-Ds\S2\PIDS\modeling_phase\Rag_gen\faiss_index_ines\index.faiss"
DATA_DIR = r"G:\Ela\study\4-Ds\S2\PIDS\modeling_phase\data\cleaned_texts"
METADATA_PATH = "./faiss_index_ines/index_metadata.pkl"
OUTPUT_DIR = r"G:\Ela\study\4-Ds\S2\PIDS\modeling_phase\Rag_gen\outputs"
DRIVE_OUT = "/content/drive/MyDrive/DATA"  # For Colab compatibility
OPENROUTER_API_KEY = os.environ.get("OPENROUTER_API_KEY", "")  # Get from environment or set directly
USE_QDRANT = True  # Set to True to use Qdrant instead of FAISS

# Ensure output directories exist
os.makedirs(DRIVE_OUT, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

class RAGEnhancedImageGenerator:
    def __init__(self):
        self.embedding_model = None
        self.sd_pipeline = None
        self.faiss_index = None
        self.metadata = None
        self.qdrant_client = None
        self.collection_name = "image_context_collection"

        # Initialize components
        self.load_embedding_model()
        self.load_stable_diffusion()
        if USE_QDRANT:
            self.setup_qdrant()
        else:
            self.load_faiss_index()

    def load_embedding_model(self):
        """Load the sentence transformer model for embeddings"""
        print("Loading embedding model...")
        self.embedding_model = SentenceTransformer(EMBEDDING_MODEL)
        print("Embedding model loaded successfully.")

    def load_stable_diffusion(self):
        """Load and optimize the Stable Diffusion pipeline"""
        print("Loading Stable Diffusion model...")
        self.sd_pipeline = StableDiffusionPipeline.from_pretrained(
            MODEL_ID,
            torch_dtype=torch.float16 if DEVICE == "cuda" else torch.float32,
            safety_checker=None,
            revision="fp16" if DEVICE == "cuda" else "main"
        )

        # Optimize the pipeline
        self.sd_pipeline.scheduler = EulerAncestralDiscreteScheduler.from_config(self.sd_pipeline.scheduler.config)

        if DEVICE == "cuda":
            self.sd_pipeline.enable_xformers_memory_efficient_attention()
            self.sd_pipeline.enable_vae_tiling()
            self.sd_pipeline.enable_attention_slicing()

        self.sd_pipeline = self.sd_pipeline.to(DEVICE)
        print("Stable Diffusion model loaded successfully.")

    def load_faiss_index(self):
        """Load the FAISS index and metadata"""
        if os.path.exists(FAISS_INDEX_PATH) and os.path.exists(METADATA_PATH):
            print("Loading FAISS index...")
            self.faiss_index = faiss.read_index(FAISS_INDEX_PATH)

            print("Loading metadata...")
            with open(METADATA_PATH, 'rb') as f:
                self.metadata = pickle.load(f)
            print(f"Loaded FAISS index with {self.faiss_index.ntotal} vectors.")
        else:
            print("FAISS index or metadata not found. Building new index...")
            self.build_faiss_index()

    def setup_qdrant(self):
        """Set up Qdrant vector database for storage and retrieval"""
        print("Setting up Qdrant client...")
        self.qdrant_client = QdrantClient(location=":memory:")  # Use disk location in production

        # Check if collection exists, if not create it
        collections = self.qdrant_client.get_collections().collections
        collection_names = [c.name for c in collections]

        if self.collection_name not in collection_names:
            print(f"Creating new Qdrant collection: {self.collection_name}")
            self.qdrant_client.create_collection(
                collection_name=self.collection_name,
                vectors_config=VectorParams(size=768, distance=Distance.COSINE),
            )

            # Load documents and populate collection
            self.populate_qdrant_collection()
        else:
            print(f"Qdrant collection {self.collection_name} already exists.")

    def populate_qdrant_collection(self):
        """Populate Qdrant collection with document embeddings"""
        print("Populating Qdrant collection with documents...")
        files = glob.glob(os.path.join(DATA_DIR, "*.txt"))

        for i, file_path in enumerate(files):
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()

            # Split into manageable chunks
            chunks = self.chunk_text(content, 512)

            for j, chunk in enumerate(chunks):
                embedding = self.embedding_model.encode(chunk)

                # Create a point with the embedding and metadata
                self.qdrant_client.upsert(
                    collection_name=self.collection_name,
                    points=[
                        PointStruct(
                            id=i * 1000 + j,
                            vector=embedding.tolist(),
                            payload={
                                "text": chunk,
                                "source": os.path.basename(file_path)
                            }
                        )
                    ]
                )

        print(f"Added {len(files)} documents to Qdrant collection.")

    def build_faiss_index(self):
        """Build a new FAISS index from documents in DATA_DIR"""
        print("Building new FAISS index...")
        files = glob.glob(os.path.join(DATA_DIR, "*.txt"))

        all_embeddings = []
        metadata = []

        for file_path in files:
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()

            # Split into manageable chunks
            chunks = self.chunk_text(content, 512)

            for chunk in chunks:
                embedding = self.embedding_model.encode(chunk)
                all_embeddings.append(embedding)
                metadata.append({
                    "text": chunk,
                    "source": os.path.basename(file_path)
                })

        # Convert to numpy array
        embeddings_array = np.array(all_embeddings).astype('float32')

        # Create FAISS index
        dimension = embeddings_array.shape[1]
        self.faiss_index = faiss.IndexFlatL2(dimension)
        self.faiss_index.add(embeddings_array)
        self.metadata = metadata

        # Save index and metadata
        os.makedirs(os.path.dirname(FAISS_INDEX_PATH), exist_ok=True)
        faiss.write_index(self.faiss_index, FAISS_INDEX_PATH)

        with open(METADATA_PATH, 'wb') as f:
            pickle.dump(metadata, f)

        print(f"Built and saved FAISS index with {len(metadata)} chunks.")

    def chunk_text(self, text: str, chunk_size: int) -> List[str]:
        """Split text into chunks of approximately chunk_size characters"""
        chunks = []
        sentences = text.split('. ')
        current_chunk = []
        current_size = 0

        for sentence in sentences:
            if current_size + len(sentence) > chunk_size and current_chunk:
                chunks.append('. '.join(current_chunk) + '.')
                current_chunk = []
                current_size = 0

            current_chunk.append(sentence)
            current_size += len(sentence)

        if current_chunk:
            chunks.append('. '.join(current_chunk) + '.')

        return chunks

    def retrieve_context(self, query: str, k: int = 3) -> List[Dict]:
        """Retrieve most relevant context from vector store"""
        query_embedding = self.embedding_model.encode(query)

        if USE_QDRANT:
            # Retrieve from Qdrant
            search_result = self.qdrant_client.search(
                collection_name=self.collection_name,
                query_vector=query_embedding.tolist(),
                limit=k
            )
            context = [{"text": hit.payload["text"], "source": hit.payload["source"], "score": hit.score}
                       for hit in search_result]
        else:
            # Retrieve from FAISS
            query_embedding = query_embedding.reshape(1, -1).astype('float32')
            distances, indices = self.faiss_index.search(query_embedding, k)

            context = []
            for i, idx in enumerate(indices[0]):
                if idx != -1 and idx < len(self.metadata):
                    item = self.metadata[idx].copy()
                    item["score"] = float(distances[0][i])
                    context.append(item)

        return context

    def enhance_prompt_with_llm(self, user_prompt: str, context: List[Dict]) -> str:
        """Enhance the user prompt using LLM with retrieved context"""
        if not OPENROUTER_API_KEY:
            print("OpenRouter API key not provided. Using original prompt.")
            return user_prompt

        # Format context for LLM input
        context_text = "\n\n".join([f"Source: {item['source']}\nContent: {item['text']}"
                                   for item in context])

        # Create system message with instructions
        system_message = """
        You are an expert at creating ultra-photorealistic financial and blockchain image generation prompts.
        Use the provided context to enhance the user's prompt for Stable Diffusion 2.1, specifically for finance, blockchain, cryptocurrency, and due diligence contexts.

        Your enhanced prompts MUST include these elements to ensure MAXIMUM PHOTOREALISM:
        1. Ultra-realistic corporate/financial setting details with precise office lighting
        2. Professional business/finance aesthetics with appropriate attire and environments
        3. Photography-first approach focusing on realism rather than artistic interpretation
        4. Specific real-world financial elements (actual trading terminal layouts, realistic monitors)
        5. Precise camera and lens specifications that would be used in professional financial photography
        6. Natural lighting conditions found in actual offices and financial institutions

        For crypto trading imagery specifically:
        - Include realistic details of actual trading environments: specific monitor brands, keyboard types
        - Reference real cryptocurrency exchange interfaces (like Binance, Coinbase Pro interfaces)
        - Show realistic charts with candlestick patterns, trading volume indicators
        - Include authentic elements like hardware wallets, smartphones with trading apps
        - Describe realistic trader appearances and environments (home office or professional trading desk)

        Formatting guide for MAXIMUM PHOTOREALISM:
        - Start with "hyper-realistic photograph of" or "professional photograph of"
        - Include "photorealistic, 8K resolution, ultra detailed, professional photography" in every prompt
        - Specify "natural lighting" or precise lighting setups found in real environments
        - Include realistic camera specs like "captured with Sony Alpha a7R IV, 24-70mm f/2.8 lens"
        - For trading scenes, include "actual cryptocurrency trading station, authentic trading terminal displays"
        - Mention "documentary-style photography" or "financial photojournalism"

        CRITICAL: Focus on creating a prompt that produces a REALISTIC PHOTOGRAPH, not digital art or illustration.
        Enhance based on context but maintain the original intent.
        Respond with only the enhanced prompt text, no explanations.
        """

        # Create user message with prompt and context
        user_message = f"""
        Original prompt: {user_prompt}

        Relevant context information:
        {context_text}

        Enhance this prompt to create a detailed, visually rich image description.
        """

        try:
            response = requests.post(
                "https://openrouter.ai/api/v1/chat/completions",
                headers={
                    "Authorization": f"Bearer {OPENROUTER_API_KEY}",
                    "Content-Type": "application/json"
                },
                json={
                    "model": MISTRAL_MODEL,
                    "messages": [
                        {"role": "system", "content": system_message},
                        {"role": "user", "content": user_message}
                    ]
                }
            )

            if response.status_code == 200:
                enhanced_prompt = response.json()["choices"][0]["message"]["content"].strip()
                print(f"Original: {user_prompt}")
                print(f"Enhanced: {enhanced_prompt}")
                return enhanced_prompt
            else:
                print(f"Error calling LLM API: {response.text}")
                return user_prompt

        except Exception as e:
            print(f"Exception when calling LLM API: {str(e)}")
            return user_prompt

    def generate_image(
        self,
        prompt: str,
        negative_prompt: str = None,
        height: int = 768,  # Increased from 512 for better details
        width: int = 768,   # Increased from 512 for better details
        num_inference_steps: int = 80,  # Increased from 60 for better quality
        guidance_scale: float = 9.0,    # Increased from 7.5 for better prompt adherence
        seed: int = None,
        use_rag: bool = True
    ) -> Image.Image:
        """Generate an image with optional RAG enhancement"""
        # Enhance prompt with RAG if requested
        if use_rag:
            # Retrieve relevant context
            context = self.retrieve_context(prompt)

            # Enhance prompt with LLM
            enhanced_prompt = self.enhance_prompt_with_llm(prompt, context)
        else:
            enhanced_prompt = prompt

        # Apply default negative prompt for financial imagery if none provided
        if negative_prompt is None or negative_prompt.strip() == "":
            negative_prompt = "cartoon, illustration, drawing, painting, sketch, anime, manga, 3d render, low quality, blurry, distorted, disfigured, deformed, watermark, signature, text, cropped, out of frame, worst quality, low quality, jpeg artifacts, poorly drawn, low detail, unrealistic, fantasy elements"

            # Add finance-specific negative terms
            if "crypto" in prompt.lower() or "trading" in prompt.lower() or "blockchain" in prompt.lower():
                negative_prompt += ", childish visualization, unrealistic screens, fake charts, simple diagram, unrealistic technology, toy-like hardware, fake interfaces, fictional diagrams, fantasy visualization"

        # Set up generator for reproducibility if seed is provided
        if seed is not None:
            generator = torch.Generator(device=DEVICE).manual_seed(seed)
        else:
            generator = None

        # Generate image with optimized parameters for realism
        out = self.sd_pipeline(
            enhanced_prompt,
            negative_prompt=negative_prompt,
            height=height,
            width=width,
            num_inference_steps=num_inference_steps,
            guidance_scale=guidance_scale,
            generator=generator,
        )

        return out.images[0], enhanced_prompt

    def save_image_with_metadata(
        self,
        image: Image.Image,
        original_prompt: str,
        enhanced_prompt: str,
        negative_prompt: str,
        seed: int,
        output_dir: str
    ) -> str:
        """Save image and metadata to specified directory"""
        # Create unique filename
        seed_str = str(seed) if seed is not None else "rand"
        timestamp = int(torch.cuda.current_device()) if DEVICE == "cuda" else 0
        filename = f"sd2-1_{seed_str}_{timestamp}"

        # Save image
        image_path = os.path.join(output_dir, f"{filename}.png")
        image.save(image_path)

        # Save metadata
        metadata = {
            "original_prompt": original_prompt,
            "enhanced_prompt": enhanced_prompt,
            "negative_prompt": negative_prompt or "",
            "seed": seed,
            "timestamp": timestamp,
            "model": MODEL_ID
        }

        metadata_path = os.path.join(output_dir, f"{filename}_metadata.json")
        with open(metadata_path, 'w') as f:
            json.dump(metadata, f, indent=2)

        return image_path
```

```python
def main():
    # Initialize the RAG-enhanced image generator
    generator = RAGEnhancedImageGenerator()

    # Get user input
    print("\n===== RAG-Enhanced Stable Diffusion 2.1 for Financial Images =====")
    user_prompt = input("Enter your image prompt: ").strip()

    # Recommend stronger negative prompt for financial imagery
    print("\nRecommended negative prompts for realistic financial imagery:")
    print("1: Default financial negative prompt (recommended)")
    print("2: Strong realistic photography negative prompt")
    print("3: Custom (enter your own)")
    neg_choice = input("Choose negative prompt (1/2/3, default=1): ").strip() or "1"

    if neg_choice == "1":
        neg_prompt = "cartoon, illustration, drawing, painting, sketch, anime, manga, 3d render, low quality, blurry, distorted, disfigured, deformed, watermark, signature, text, cropped, out of frame, worst quality, low quality, jpeg artifacts, poorly drawn, low detail, unrealistic, fantasy elements, childish visualization, unrealistic screens, fake charts, simple diagram, unrealistic technology, toy-like hardware"
    elif neg_choice == "2":
        neg_prompt = "cartoon, illustration, drawing, painting, 3d render, digital art, anime, manga, sketch, watercolor, comic, rendered, CG, CGI, drawing, painting, crayon, sketch, graphite, impressionist, noisy, blurry, soft, deformed, distorted, disfigured, poorly drawn face, mutation, mutated, extra limb, ugly, poorly drawn hands, missing limb, floating limbs, disconnected limbs, malformed hands, blurry, watermark, watermarked, oversaturated, censored, amateur drawing, signature, low contrast, underexposed, overexposed, beginner, amateur, low resolution"
    else:
        neg_prompt = input("Enter custom negative prompt: ").strip() or None

    seed_input = input("Seed (blank for random): ").strip()
    use_rag = input("Use RAG enhancement? (y/n, default: y): ").strip().lower() != "n"

    # Get resolution preference
    res_choice = input("Resolution (1: 768x768, 2: 1024x1024, 3: Custom, default=1): ").strip() or "1"
    if res_choice == "1":
        height, width = 768, 768
    elif res_choice == "2":
        height, width = 1024, 1024
    else:
        height = int(input("Enter height (512-1024, default=768): ").strip() or "768")
        width = int(input("Enter width (512-1024, default=768): ").strip() or "768")

    seed = int(seed_input) if seed_input.isdigit() else None

    if not user_prompt:
        print("No prompt—exiting.")
        exit()

    # Generate image with improved parameters
    print(f"Generating ultra-realistic image{' with RAG enhancement' if use_rag else ''}...")
    print(f"Using {height}x{width} resolution with enhanced quality settings")

    image, enhanced_prompt = generator.generate_image(
        prompt=user_prompt,
        negative_prompt=neg_prompt,
        height=height,
        width=width,
        num_inference_steps=80,  # Higher step count for better quality
        guidance_scale=9.0,      # Higher guidance for better prompt adherence
        seed=seed,
        use_rag=use_rag
    )

    # Save image and metadata
    output_path = generator.save_image_with_metadata(
        image=image,
        original_prompt=user_prompt,
        enhanced_prompt=enhanced_prompt,
        negative_prompt=neg_prompt,
        seed=seed,
        output_dir=OUTPUT_DIR
    )

    # Also save to DRIVE_OUT for Colab compatibility
    colab_path = os.path.join(DRIVE_OUT, os.path.basename(output_path))
    image.save(colab_path)

    print(f"✅ Saved to {output_path}")
    if DRIVE_OUT != OUTPUT_DIR:
        print(f"✅ Also saved to {colab_path}")

    # Display enhancement info
    if use_rag and enhanced_prompt != user_prompt:
        print("\n=== Prompt Enhancement for Maximum Realism ===")
        print(f"Original: {user_prompt}")
        print(f"Enhanced: {enhanced_prompt}")

    # Offer to generate another with different settings
    another = input("\nGenerate another image with different settings? (y/n, default=n): ").strip().lower() == "y"
    if another:
        print("\nStarting new generation...")
        main()  # Recursive call to start again

if __name__ == "__main__":
    main()
```
