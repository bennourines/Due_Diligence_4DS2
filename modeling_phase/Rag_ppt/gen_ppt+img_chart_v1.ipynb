```python
# @title
# Cell 1: Dependencies and Utility Functions
import os
import pickle
import torch
import numpy as np
import pandas as pd
import re
import tempfile
import urllib.request
import random
import json
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from typing import List, Dict, Any, Tuple, Optional, Union

# Import libraries for image generation
from diffusers import (
    StableDiffusionPipeline,
    DPMSolverMultistepScheduler,
    EulerAncestralDiscreteScheduler
)
from PIL import Image, ImageDraw, ImageFont

# Import libraries for presentation generation
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
from pptx.chart.data import ChartData
from pptx.enum.chart import XL_CHART_TYPE

# Import libraries for forecasting
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import scipy.stats as stats

# Optional RAG-related imports
try:
    from sentence_transformers import SentenceTransformer
    import faiss
    from qdrant_client import QdrantClient
    from qdrant_client.models import Distance, VectorParams, PointStruct
    VECTOR_SEARCH_AVAILABLE = True
except ImportError:
    VECTOR_SEARCH_AVAILABLE = False
    print("Vector search libraries not available. Running without RAG enhancement.")
```

```python


# Configuration
MODEL_ID = "stabilityai/stable-diffusion-2-1"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
EMBEDDING_MODEL = "multi-qa-mpnet-base-dot-v1" if VECTOR_SEARCH_AVAILABLE else None
COLAB_DRIVE_PATH = "/content/drive/MyDrive/PPT_Generator"
USE_QDRANT = True  # Use in-memory faiss by default in Colab

# Create directories for outputs
os.makedirs(COLAB_DRIVE_PATH, exist_ok=True)
os.makedirs(f"{COLAB_DRIVE_PATH}/images", exist_ok=True)
os.makedirs(f"{COLAB_DRIVE_PATH}/presentations", exist_ok=True)
os.makedirs(f"{COLAB_DRIVE_PATH}/charts", exist_ok=True)

# RAG-Enhanced Image Generator Class
class RAGEnhancedImageGenerator:
    def __init__(self):
        self.embedding_model = None
        self.sd_pipeline = None
        self.faiss_index = None
        self.metadata = None
        self.qdrant_client = None
        self.collection_name = "image_context_collection"

        # Initialize components
        if VECTOR_SEARCH_AVAILABLE:
            self.load_embedding_model()
        self.load_stable_diffusion()

    def load_embedding_model(self):
        """Load the sentence transformer model for embeddings"""
        print("Loading embedding model...")
        self.embedding_model = SentenceTransformer(EMBEDDING_MODEL)
        print("Embedding model loaded successfully.")

    def load_stable_diffusion(self):
        """Load and optimize the Stable Diffusion pipeline"""
        print("Loading Stable Diffusion model...")
        self.sd_pipeline = StableDiffusionPipeline.from_pretrained(
            MODEL_ID,
            torch_dtype=torch.float16 if DEVICE == "cuda" else torch.float32,
            safety_checker=None,
            revision="fp16" if DEVICE == "cuda" else "main"
      )

        # Optimize the pipeline
        self.sd_pipeline.scheduler = EulerAncestralDiscreteScheduler.from_config(self.sd_pipeline.scheduler.config)

        if DEVICE == "cuda":
            # Check if xformers is available before enabling
            try:
                import xformers
                if hasattr(self.sd_pipeline, "enable_xformers_memory_efficient_attention"):
                    self.sd_pipeline.enable_xformers_memory_efficient_attention()
            except ImportError:
                print("xformers not available, skipping memory efficient attention")

        # These optimizations should work regardless of xformers
            if hasattr(self.sd_pipeline, "enable_vae_tiling"):
                self.sd_pipeline.enable_vae_tiling()
            self.sd_pipeline.enable_attention_slicing()

        self.sd_pipeline = self.sd_pipeline.to(DEVICE)
        print("Stable Diffusion model loaded successfully.")

    def generate_image(
        self,
        prompt: str,
        negative_prompt: str = None,
        height: int = 768,
        width: int = 768,
        num_inference_steps: int = 50,
        guidance_scale: float = 8.5,
        seed: int = None,
    ) -> Tuple[Image.Image, str]:
        """Generate an image based on the prompt"""
        # Default negative prompt for professional/realistic images
        if negative_prompt is None:
            negative_prompt = "cartoon, illustration, drawing, painting, sketch, anime, manga, 3d render, low quality, blurry, distorted, disfigured, deformed, watermark, signature, text, cropped, out of frame, worst quality, low quality, jpeg artifacts, poorly drawn, low detail"

        # Enhance the prompt for better results (specialized for slide content)
        enhanced_prompt = self._enhance_prompt_for_slides(prompt)

        # Set up generator for reproducibility if seed is provided
        if seed is not None:
            generator = torch.Generator(device=DEVICE).manual_seed(seed)
        else:
            generator = None

        # Generate image
        out = self.sd_pipeline(
            enhanced_prompt,
            negative_prompt=negative_prompt,
            height=height,
            width=width,
            num_inference_steps=num_inference_steps,
            guidance_scale=guidance_scale,
            generator=generator,
        )

        return out.images[0], enhanced_prompt

    def _enhance_prompt_for_slides(self, prompt: str) -> str:
        """Enhance the prompt to create better slide images"""
        # Basic prompt improvement for slides
        prompt_prefix = "professional high quality photograph of "
        prompt_suffix = ", corporate style, business presentation, professional lighting, clear focus, 8K resolution"

        # Don't duplicate prefixes
        if not prompt.lower().startswith(prompt_prefix.lower()):
            enhanced_prompt = f"{prompt_prefix}{prompt}{prompt_suffix}"
        else:
            enhanced_prompt = f"{prompt}{prompt_suffix}"

        return enhanced_prompt

    def save_image(self, image: Image.Image, title: str, output_dir: str = f"{COLAB_DRIVE_PATH}/images") -> str:
        """Save generated image to file"""
        # Create a safe filename
        safe_title = re.sub(r'[^\w\-]', '_', title.lower().strip())
        filename = f"{safe_title}_{random.randint(1000, 9999)}.png"
        filepath = os.path.join(output_dir, filename)

        # Save the image
        image.save(filepath)
        print(f"Saved image to {filepath}")

        return filepath

# Chart Generation Utilities
class ChartGenerator:
    @staticmethod
    def generate_time_series_data(
        start_date: str,
        periods: int = 24,
        trend: float = 0.1,
        seasonality: float = 0.2,
        noise: float = 0.05,
        forecast_periods: int = 6
    ) -> Tuple[pd.DataFrame, pd.DataFrame]:
        """Generate synthetic time series data with trend, seasonality, and forecast"""
        # Parse start date
        start = datetime.strptime(start_date, "%Y-%m-%d")

        # Generate dates
        dates = [start + timedelta(days=30*i) for i in range(periods)]

        # Generate values with trend, seasonality, and noise
        values = []
        for i in range(periods):
            trend_component = i * trend
            seasonal_component = seasonality * np.sin(2 * np.pi * i / 12)  # Annual seasonality
            noise_component = noise * np.random.randn()
            value = 100 + trend_component + seasonal_component + noise_component
            values.append(max(0, value))  # Ensure no negative values

        # Create historical dataframe
        historical_df = pd.DataFrame({
            'Date': dates,
            'Value': values
        })

        # Generate forecast
        forecast_dates = [dates[-1] + timedelta(days=30*(i+1)) for i in range(forecast_periods)]
        forecast_values = []

        # Simple forecasting model (could be replaced with ARIMA or other methods)
        for i in range(forecast_periods):
            idx = periods + i
            trend_component = idx * trend
            seasonal_component = seasonality * np.sin(2 * np.pi * idx / 12)
            forecast_value = 100 + trend_component + seasonal_component
            forecast_values.append(max(0, forecast_value))

        # Create forecast dataframe
        forecast_df = pd.DataFrame({
            'Date': forecast_dates,
            'Value': forecast_values
        })

        return historical_df, forecast_df

    @staticmethod
    def generate_arima_forecast(
        historical_data: pd.DataFrame,
        periods: int = 6,
        order: Tuple[int, int, int] = (1, 1, 1)
    ) -> pd.DataFrame:
        """Generate ARIMA forecast from historical data"""
        # Fit ARIMA model
        model = ARIMA(historical_data['Value'], order=order)
        model_fit = model.fit()

        # Make forecast
        forecast = model_fit.forecast(steps=periods)

        # Generate forecast dates
        last_date = historical_data['Date'].iloc[-1]
        forecast_dates = [last_date + timedelta(days=30*(i+1)) for i in range(periods)]

        # Create forecast dataframe
        forecast_df = pd.DataFrame({
            'Date': forecast_dates,
            'Value': forecast.values
        })

        return forecast_df

    @staticmethod
    def create_chart_figure(
        historical_data: pd.DataFrame,
        forecast_data: pd.DataFrame,
        title: str,
        y_label: str,
        chart_type: str = 'line'
    ) -> str:
        """Create and save a chart figure"""
        plt.figure(figsize=(10, 6))

        # Plot historical data
        if chart_type == 'line':
            plt.plot(historical_data['Date'], historical_data['Value'], 'b-', label='Historical')
            plt.plot(forecast_data['Date'], forecast_data['Value'], 'r--', label='Forecast')
        elif chart_type == 'bar':
            plt.bar(historical_data['Date'], historical_data['Value'], color='blue', alpha=0.7, label='Historical')
            plt.bar(forecast_data['Date'], forecast_data['Value'], color='red', alpha=0.7, label='Forecast')

        # Add labels and title
        plt.xlabel('Date')
        plt.ylabel(y_label)
        plt.title(title)
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.xticks(rotation=45)
        plt.tight_layout()

        # Save figure
        safe_title = re.sub(r'[^\w\-]', '_', title.lower().strip())
        filepath = os.path.join(f"{COLAB_DRIVE_PATH}/charts", f"{safe_title}_{random.randint(1000, 9999)}.png")
        plt.savefig(filepath, dpi=300)
        plt.close()

        return filepath

    @staticmethod
    def generate_chart_data_for_ppt(
        title: str,
        time_period: str = "Monthly",
        chart_type: str = "column",
        topic: str = None,
        forecast: bool = True
    ) -> Dict[str, Any]:
        """Generate chart data suitable for PowerPoint integration"""

        # Generate appropriate labels and data based on topic
        if topic and "blockchain" in topic.lower():
            labels = ["Bitcoin", "Ethereum", "Cardano", "Solana", "Polkadot"]
            values = [random.randint(30, 100) for _ in range(len(labels))]
            chart_info = {
                "type": chart_type,
                "title": title,
                "data": {
                    "labels": labels,
                    "values": values
                }
            }

        elif topic and ("finance" in topic.lower() or "market" in topic.lower()):
            # Generate time series data for financial markets
            start_date = "2023-01-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=12,
                forecast_periods=4 if forecast else 0
            )

            # Create labels and values from the data
            dates = [d.strftime("%b %Y") for d in historical_df['Date']]
            values = [round(v, 2) for v in historical_df['Value']]

            if forecast:
                forecast_dates = [d.strftime("%b %Y") for d in forecast_df['Date']]
                forecast_values = [round(v, 2) for v in forecast_df['Value']]

                # Create chart data with two series
                chart_info = {
                    "type": "line" if chart_type == "line" else "column",
                    "title": title,
                    "data": {
                        "labels": dates + forecast_dates,
                        "series": [
                            {
                                "name": "Historical",
                                "values": values + [None] * len(forecast_dates)
                            },
                            {
                                "name": "Forecast",
                                "values": [None] * len(dates) + forecast_values
                            }
                        ]
                    }
                }
            else:
                chart_info = {
                    "type": chart_type,
                    "title": title,
                    "data": {
                        "labels": dates,
                        "values": values
                    }
                }
        else:
            # Generic time series for any other topic
            if time_period == "Quarterly":
                labels = ["Q1 2023", "Q2 2023", "Q3 2023", "Q4 2023", "Q1 2024", "Q2 2024"]
            elif time_period == "Yearly":
                labels = ["2019", "2020", "2021", "2022", "2023", "2024"]
            else:  # Monthly
                labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

            # Generate semi-random values with upward trend
            base = 40
            values = [max(10, base + i*3 + random.randint(-10, 15)) for i in range(len(labels))]

            chart_info = {
                "type": chart_type,
                "title": title,
                "data": {
                    "labels": labels,
                    "values": values
                }
            }

        return chart_info

# Presentation Utility Functions
def create_slide_with_image(
    presentation: Presentation,
    title: str,
    image_path: str,
    caption: Optional[str] = None
) -> None:
    """Add a slide with an image to the presentation"""
    # Add a new slide
    slide_layout = presentation.slide_layouts[5]  # Title and Content
    slide = presentation.slides.add_slide(slide_layout)

    # Set the title
    slide.shapes.title.text = title

    # Add the image
    if os.path.exists(image_path):
        left = Inches(1.5)
        top = Inches(2)
        width = Inches(7)  # Fixed width
        picture = slide.shapes.add_picture(image_path, left, top, width=width)

        # Add caption if provided
        if caption:
            left = Inches(1.5)
            top = Inches(5.5)
            width = Inches(7)
            height = Inches(0.5)
            textbox = slide.shapes.add_textbox(left, top, width, height)
            textbox.text_frame.text = caption
            textbox.text_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
    else:
        print(f"Warning: Image not found at {image_path}")

def create_chart_slide(
    presentation: Presentation,
    title: str,
    chart_info: Dict[str, Any]
) -> None:
    """Add a slide with a chart based on provided data"""
    # Add a new slide
    slide = presentation.slides.add_slide(presentation.slide_layouts[5])  # Title and Content
    slide.shapes.title.text = title

    # Prepare chart data
    chart_data = ChartData()

    # Handle different data formats
    if "series" in chart_info.get("data", {}):
        # Multiple series chart
        series_data = chart_info["data"]["series"]
        chart_data.categories = chart_info["data"]["labels"]

        for series in series_data:
            chart_data.add_series(series["name"], series["values"])
    else:
        # Single series chart
        labels = chart_info.get("data", {}).get("labels", [])
        values = chart_info.get("data", {}).get("values", [])

        if not labels or not values:
            print(f"Warning: No chart data for {title}")
            return

        chart_data.categories = labels
        chart_data.add_series(chart_info.get("title", "Series 1"), values)

    # Map chart type
    type_map = {
        "bar": XL_CHART_TYPE.BAR_CLUSTERED,
        "column": XL_CHART_TYPE.COLUMN_CLUSTERED,
        "line": XL_CHART_TYPE.LINE_MARKERS,
        "pie": XL_CHART_TYPE.PIE,
        "area": XL_CHART_TYPE.AREA
    }
    chart_type = type_map.get(chart_info.get("type", "column"), XL_CHART_TYPE.COLUMN_CLUSTERED)

    # Insert chart
    x, y = Inches(1), Inches(2)
    cx, cy = Inches(8), Inches(4.5)
    chart_frame = slide.shapes.add_chart(chart_type, x, y, cx, cy, chart_data)

    # Set chart title
    chart = chart_frame.chart
    chart.has_title = True
    chart.chart_title.text_frame.text = chart_info.get("title", "")

def create_content_slide(
    presentation: Presentation,
    title: str,
    content: List[str]
) -> None:
    """Create a slide with bullet points"""
    # Add a new slide
    slide = presentation.slides.add_slide(presentation.slide_layouts[1])  # Title and Content

    # Set the title
    slide.shapes.title.text = title

    # Add bullet points
    text_frame = slide.placeholders[1].text_frame
    text_frame.clear()

    for i, point in enumerate(content):
        p = text_frame.add_paragraph()
        p.text = point
        p.level = 0

def create_title_slide(
    presentation: Presentation,
    title: str,
    subtitle: str = None
) -> None:
    """Create a title slide"""
    # Add title slide
    slide = presentation.slides.add_slide(presentation.slide_layouts[0])  # Title Slide

    # Set title
    slide.shapes.title.text = title

    # Set subtitle if provided
    if subtitle and len(slide.placeholders) > 1:
        slide.placeholders[1].text = subtitle
```

```python
# Cell 2: Integrated PowerPoint Generator with AI Images and Forecast Charts

class IntegratedPresentationGenerator:
    """
    Class that integrates presentation generation with AI image generation
    and time series forecasting for charts
    """

    def __init__(self):
        """Initialize the integrated presentation generator"""
        # Initialize the image generator
        print("Initializing AI image generator...")
        self.image_generator = RAGEnhancedImageGenerator()

        # Output paths
        self.output_dir = f"{COLAB_DRIVE_PATH}/presentations"
        self.image_dir = f"{COLAB_DRIVE_PATH}/images"
        self.chart_dir = f"{COLAB_DRIVE_PATH}/charts"

        # Ensure directories exist
        os.makedirs(self.output_dir, exist_ok=True)
        os.makedirs(self.image_dir, exist_ok=True)
        os.makedirs(self.chart_dir, exist_ok=True)

    def generate_presentation_structure(self, topic: str, num_slides: int = 7) -> Dict[str, Any]:
        """
        Generate a structured presentation based on the topic

        Args:
            topic: The presentation topic
            num_slides: Desired number of slides

        Returns:
            Dictionary containing the presentation structure
        """
        print(f"Generating presentation structure for topic: {topic}")

        # Create the standard slide templates for different topics
        templates = {
            # General business presentation
            "general": [
                {"title": "Introduction", "content": [
                    f"Overview of {topic}",
                    "Key objectives and goals",
                    "Importance in the current context"
                ]},
                {"title": "Market Overview", "content": [
                    "Current market landscape",
                    "Key market trends and developments",
                    "Market size and growth projections"
                ]},
                {"title": "Key Components", "content": [
                    f"Core elements of {topic}",
                    "Important features and technologies",
                    "Integration points with existing systems"
                ]},
                {"title": "Data Analysis", "content": [
                    "Historical performance metrics",
                    "Comparative analysis with industry benchmarks",
                    "Key insights from data"
                ]},
                {"title": "Future Trends", "content": [
                    f"Emerging trends in {topic}",
                    "Short-term predictions (1-2 years)",
                    "Long-term forecast (3-5 years)"
                ]},
                {"title": "Case Studies", "content": [
                    "Real-world implementation examples",
                    "Success stories and outcomes",
                    "Lessons learned from implementations"
                ]},
                {"title": "Implementation Strategy", "content": [
                    "Recommended approach for implementation",
                    "Timeline and key milestones",
                    "Resource requirements and considerations"
                ]},
                {"title": "Risks and Mitigation", "content": [
                    "Potential challenges and risks",
                    "Mitigation strategies",
                    "Contingency planning"
                ]},
                {"title": "Summary", "content": [
                    f"Key takeaways about {topic}",
                    "Next steps and recommendations",
                    "Questions and discussion"
                ]}
            ],

            # Blockchain specific presentation
            "blockchain": [
                {"title": "Introduction to Blockchain", "content": [
                    "What is blockchain technology",
                    "Key principles: decentralization, immutability, and transparency",
                    "Evolution of blockchain technology"
                ]},
                {"title": "Blockchain Architecture", "content": [
                    "Core components of blockchain systems",
                    "Consensus mechanisms (PoW, PoS, DPoS)",
                    "Smart contracts and decentralized applications"
                ]},
                {"title": "Market Analysis", "content": [
                    "Current blockchain market overview",
                    "Key players and platforms",
                    "Market size and growth projections"
                ]},
                {"title": "Use Cases", "content": [
                    "Financial services applications",
                    "Supply chain management",
                    "Healthcare and identity management",
                    "Government and public sector applications"
                ]},
                {"title": "Cryptocurrency Trends", "content": [
                    "Major cryptocurrencies performance",
                    "Trading volume and market capitalization",
                    "Regulatory landscape and compliance"
                ]},
                {"title": "Technical Challenges", "content": [
                    "Scalability solutions",
                    "Energy consumption concerns",
                    "Interoperability between blockchains",
                    "Security considerations"
                ]},
                {"title": "Implementation Framework", "content": [
                    "Blockchain selection criteria",
                    "Integration approaches with existing systems",
                    "Development and deployment considerations"
                ]},
                {"title": "Future of Blockchain", "content": [
                    "Emerging trends and innovations",
                    "Potential disruptive applications",
                    "Long-term industry forecast"
                ]},
                {"title": "Summary", "content": [
                    "Key takeaways from blockchain analysis",
                    "Strategic recommendations",
                    "Next steps and action plan"
                ]}
            ],

            # Financial analysis presentation
            "finance": [
                {"title": "Financial Overview", "content": [
                    "Key financial metrics and highlights",
                    "Year-over-year performance",
                    "Industry benchmarking"
                ]},
                {"title": "Market Analysis", "content": [
                    "Market conditions and economic factors",
                    "Competitive landscape",
                    "Market share analysis"
                ]},
                {"title": "Revenue Analysis", "content": [
                    "Revenue streams breakdown",
                    "Growth trends by segment",
                    "Quarterly and annual comparisons"
                ]},
                {"title": "Cost Structure", "content": [
                    "Operating expenses analysis",
                    "Cost optimization opportunities",
                    "Fixed vs. variable costs"
                ]},
                {"title": "Profitability Metrics", "content": [
                    "Gross margin, EBITDA, and net profit",
                    "Profit center analysis",
                    "Profitability improvement strategies"
                ]},
                {"title": "Cash Flow Analysis", "content": [
                    "Operating, investing, and financing cash flows",
                    "Cash flow forecasting",
                    "Working capital management"
                ]},
                {"title": "Investment Analysis", "content": [
                    "ROI on current investments",
                    "Capital allocation strategy",
                    "Investment opportunities evaluation"
                ]},
                {"title": "Financial Forecast", "content": [
                    "Short-term financial projections",
                    "Long-term financial outlook",
                    "Sensitivity analysis and scenarios"
                ]},
                {"title": "Summary and Recommendations", "content": [
                    "Key financial insights",
                    "Strategic financial recommendations",
                    "Action plan and next steps"
                ]}
            ]
        }

        # Determine which template to use based on the topic
        if "blockchain" in topic.lower() or "crypto" in topic.lower() or "bitcoin" in topic.lower():
            selected_template = templates["blockchain"]
        elif "financ" in topic.lower() or "investment" in topic.lower() or "market" in topic.lower():
            selected_template = templates["finance"]
        else:
            selected_template = templates["general"]

        # Adjust the number of slides
        max_slides = len(selected_template)
        if num_slides > max_slides:
            num_slides = max_slides
            print(f"Limiting to maximum available slides: {num_slides}")

        # Select slides (always include intro and summary)
        if num_slides <= 2:
            selected_slides = [selected_template[0], selected_template[-1]]
        else:
            # Include first and last slide, select middle ones
            middle_slides = selected_template[1:-1]
            if num_slides - 2 < len(middle_slides):
                # Randomly select slides to remove
                slides_to_remove = len(middle_slides) - (num_slides - 2)
                indices_to_remove = random.sample(range(len(middle_slides)), slides_to_remove)
                middle_slides = [s for i, s in enumerate(middle_slides) if i not in indices_to_remove]

            selected_slides = [selected_template[0]] + middle_slides + [selected_template[-1]]

        # Create the presentation structure
        presentation = {
            "title": f"{topic} - Comprehensive Analysis",
            "subtitle": "Strategic Insights and Forecast",
            "slides": selected_slides
        }

        return presentation

    def generate_image_for_slide(self, slide_title: str, slide_content: List[str]) -> str:
        """
        Generate an AI image that directly relates to the slide content

        Args:
            slide_title: The slide title
            slide_content: List of bullet points for the slide

        Returns:
            Path to the generated image
        """
        print(f"Generating image for slide: {slide_title}")

        # Create a descriptive prompt based on slide content
        content_text = " ".join(slide_content)
        # Keep prompt focused and not too long
        prompt = f"{slide_title}: {content_text[:100]}"

        try:
            # Generate the image
            image, enhanced_prompt = self.image_generator.generate_image(
                prompt=prompt,
                height=768,
                width=1024,
                num_inference_steps=50,
                guidance_scale=8.5
            )

            # Save the image
            image_path = self.image_generator.save_image(image, slide_title, self.image_dir)
            print(f"  Image prompt: {enhanced_prompt}")

            return image_path
        except Exception as e:
            print(f"  Error generating image: {str(e)}")
            # Create a placeholder image with text instead
            return self.create_placeholder_image(slide_title, slide_content)

    def create_placeholder_image(self, title: str, content: List[str]) -> str:
        """Create a placeholder image with text when AI generation fails"""
        width, height = 1024, 768
        img = Image.new('RGB', (width, height), color=(240, 240, 245))
        draw = ImageDraw.Draw(img)

        # Try to use a font, or fall back to default
        try:
            # Try some common fonts
            font_path = None
            try:
                from google.colab import auth
                # In Colab, use default font
                title_font = ImageFont.load_default()
                text_font = ImageFont.load_default()
            except:
                # Try system fonts
                font_dirs = ["/usr/share/fonts/truetype/", "/Windows/Fonts/", "/System/Library/Fonts/"]
                font_files = ["arial.ttf", "Arial.ttf", "DejaVuSans.ttf", "Verdana.ttf"]

                for font_dir in font_dirs:
                    for font_file in font_files:
                        if os.path.exists(os.path.join(font_dir, font_file)):
                            font_path = os.path.join(font_dir, font_file)
                            break
                    if font_path:
                        break

                if font_path:
                    title_font = ImageFont.truetype(font_path, 40)
                    text_font = ImageFont.truetype(font_path, 30)
                else:
                    title_font = ImageFont.load_default()
                    text_font = ImageFont.load_default()
        except:
            title_font = ImageFont.load_default()
            text_font = ImageFont.load_default()

        # Draw title
        draw.text((width/2, 100), title, fill=(30, 30, 90), font=title_font, anchor="mt")

        # Draw content as bullet points
        y_pos = 200
        for item in content:
            if len(item) > 60:
                item = item[:57] + "..."
            draw.text((width/4, y_pos), f"• {item}", fill=(60, 60, 60), font=text_font)
            y_pos += 60

        # Save image
        safe_title = re.sub(r'[^\w\-]', '_', title.lower().strip())
        filepath = os.path.join(self.image_dir, f"{safe_title}_placeholder.png")
        img.save(filepath)

        return filepath

    def generate_chart_for_slide(self, slide_title: str, slide_content: List[str], presentation_topic: str) -> str:
        """
        Generate a relevant chart for the slide with forecasting elements

        Args:
            slide_title: The slide title
            slide_content: List of bullet points for the slide
            presentation_topic: Overall presentation topic for context

        Returns:
            Path to the generated chart image
        """
        print(f"Generating chart for slide: {slide_title}")

        # Determine chart type based on slide content
        chart_type = "column"  # default
        content_text = " ".join(slide_content).lower()

        if any(word in content_text for word in ["trend", "forecast", "projection", "growth", "future"]):
            chart_type = "line"  # Line charts for trends and forecasts
        elif any(word in content_text for word in ["breakdown", "distribution", "allocation", "composition"]):
            chart_type = "pie"   # Pie charts for distributions
        elif any(word in content_text for word in ["comparison", "versus", "contrast"]):
            chart_type = "bar"   # Bar charts for comparisons

        # Generate chart data with forecasting elements
        needs_forecast = any(word in content_text for word in ["forecast", "projection", "future", "predict"])

        # Generate specialized chart based on content
        if "revenue" in content_text or "sales" in content_text:
            # Generate revenue/sales forecast
            start_date = "2022-01-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=12,  # 1 year of data
                trend=0.8,   # Upward trend
                seasonality=0.3,
                forecast_periods=4 if needs_forecast else 0
            )

            # Create chart image
            chart_title = f"{slide_title} - {'With Forecast' if needs_forecast else 'Historical'}"
            chart_path = ChartGenerator.create_chart_figure(
                historical_data=historical_df,
                forecast_data=forecast_df,
                title=chart_title,
                y_label="Revenue ($K)" if "revenue" in content_text else "Sales Volume",
                chart_type=chart_type
            )

        elif "profit" in content_text or "margin" in content_text:
            # Generate profit/margin forecast with more variability
            start_date = "2022-01-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=12,
                trend=0.5,   # Moderate trend
                seasonality=0.4,  # More seasonal variation
                noise=0.1,    # More noise
                forecast_periods=4 if needs_forecast else 0
            )

            # Create chart image
            chart_title = f"{slide_title} - {'With Forecast' if needs_forecast else 'Historical'}"
            chart_path = ChartGenerator.create_chart_figure(
                historical_data=historical_df,
                forecast_data=forecast_df,
                title=chart_title,
                y_label="Profit Margin (%)" if "margin" in content_text else "Profit ($K)",
                chart_type=chart_type
            )

        elif "market" in content_text or "industry" in content_text:
            # Generate market/industry forecast with more conservative trend
            start_date = "2021-01-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=18,  # 1.5 years of data
                trend=0.3,   # Conservative trend
                seasonality=0.2,
                forecast_periods=6 if needs_forecast else 0  # Longer forecast
            )

            # Create chart image
            chart_title = f"{slide_title} - {'With Forecast' if needs_forecast else 'Historical'}"
            chart_path = ChartGenerator.create_chart_figure(
                historical_data=historical_df,
                forecast_data=forecast_df,
                title=chart_title,
                y_label="Market Size ($B)" if "market" in content_text else "Industry Growth (%)",
                chart_type=chart_type
            )

        elif "crypto" in content_text or "blockchain" in presentation_topic.lower():
            # Generate cryptocurrency price chart with high volatility
            start_date = "2021-06-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=24,  # 2 years of data
                trend=0.1,   # Slight trend
                seasonality=0.5,  # High seasonality
                noise=0.2,   # High volatility/noise
                forecast_periods=6 if needs_forecast else 0
            )

            # Create chart image
            chart_title = f"Cryptocurrency Performance {'& Projection' if needs_forecast else ''}"
            chart_path = ChartGenerator.create_chart_figure(
                historical_data=historical_df,
                forecast_data=forecast_df,
                title=chart_title,
                y_label="Price (USD)",
                chart_type="line"  # Always use line for crypto
            )

        else:
            # Generic forecast chart
            start_date = "2022-01-01"
            historical_df, forecast_df = ChartGenerator.generate_time_series_data(
                start_date=start_date,
                periods=12,
                forecast_periods=4 if needs_forecast else 0
            )

            # Create chart image
            chart_title = f"{slide_title} - {'Forecast Analysis' if needs_forecast else 'Data Analysis'}"
            chart_path = ChartGenerator.create_chart_figure(
                historical_data=historical_df,
                forecast_data=forecast_df,
                title=chart_title,
                y_label="Value",
                chart_type=chart_type
            )

        return chart_path

    def create_presentation(self, topic: str, num_slides: int = 7) -> str:
        """
        Generate a complete presentation with content, AI images, and forecast charts

        Args:
            topic: The presentation topic
            num_slides: Desired number of slides

        Returns:
            Path to the generated PowerPoint file
        """
        print(f"Creating presentation on '{topic}' with {num_slides} slides...")

        # Generate presentation structure
        content = self.generate_presentation_structure(topic, num_slides)

        # Create PowerPoint presentation
        prs = Presentation()

        # Add title slide
        create_title_slide(prs, content["title"], content["subtitle"])

        # Process each content slide
        for slide_data in content["slides"]:
            slide_title = slide_data["title"]
            slide_content = slide_data["content"]

            # 1. Create content slide with bullet points
            create_content_slide(prs, slide_title, slide_content)

            # 2. Generate and add image slide based on content
            image_path = self.generate_image_for_slide(slide_title, slide_content)
            create_slide_with_image(prs, f"{slide_title} - Visual", image_path)

            # 3. Generate and add chart slide based on content
            chart_path = self.generate_chart_for_slide(slide_title, slide_content, topic)
            create_slide_with_image(prs, f"{slide_title} - Analytics", chart_path,
                                   caption="Data analysis with historical trends and forecast")

        # Save the presentation
        #safe_topic = re.sub(r'[^\w\- ]', '_', topic).strip().replace(' ', '_')[:50] #Change this line to truncate topic

        output_path = os.path.join(self.output_dir, "due deal LST presentation.pptx") # Changed line
        prs.save(output_path)

        print(f"✅ Presentation saved to: {output_path}")
        return output_path

# Main execution function
def generate_presentation():
    """Interactive function to generate a presentation"""
    # Get user input
    print("=" * 70)
    print("🌟 AI-Powered Presentation Generator with Images & Forecast Charts 🌟")
    print("=" * 70)

    topic = input("\n📝 Enter presentation topic: ")
    num_slides_input = input("🔢 Number of slides (default: 7): ").strip()
    num_slides = int(num_slides_input) if num_slides_input.isdigit() else 7

    # Create generator and generate presentation
    generator = IntegratedPresentationGenerator()
    output_path = generator.create_presentation(topic, num_slides)

    # Display results
    print("\n✅ Generation complete!")
    print(f"📊 Presentation saved to: {output_path}")
    print(f"🖼️ Images saved to: {generator.image_dir}")
    print(f"📈 Charts saved to: {generator.chart_dir}")

    return output_path

# Execute when run
if __name__ == "__main__":
    presentation_path = generate_presentation()

# Call the function to generate a presentation
presentation_path = generate_presentation()
```
